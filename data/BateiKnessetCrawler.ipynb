```python
import re
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def get_cities(driver):
    """
    Get list of cities from the website
    """
    url = "https://www.kipa.co.il/%D7%91%D7%AA%D7%99-%D7%9B%D7%A0%D7%A1%D7%AA/%D7%91%D7%AA%D7%99-%D7%9B%D7%A0%D7%A1%D7%AA-%D7%9C%D7%A4%D7%99-%D7%A2%D7%99%D7%A8/"
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    cities = []
    for city in soup.find_all('ul',{'style':'  column-count: 3;  column-gap: 40px;'}):
        for city_name in city.find_all('li'):
            cities.append(city_name.text)
    driver.quit()
    return cities
```

```python
def get_nosahim(cities,driver):
    
    output_rows = []
    driver = webdriver.Chrome(ChromeDriverManager().install())
    for city in cities:
        print('getting {}'.format(city))
        url = "https://www.kipa.co.il/בתי-כנסת/{}".format(city)
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, "html.parser")
        table = soup.find("table")
        if table is not None:
            for table_row in table.find_all('tr'):
                columns = table_row.find_all('td')
                output_row = [city]
                for column in columns:
                    output_row.append(column.text)
                if len(output_row) > 2: ## If there is a row with no data
                    if "," in output_row[3]:
                        if re.search(r'[0-9]+', output_row[3]): ## If there is a number in the address
                            output_row[3] = output_row[3].split(",")[0].rstrip()
                            output_rows.append(output_row)

    with open('/Users/daniel.ofir/Documents/Programming Projects/CS/Data Science/Pray_Around/batei.csv', 'w', encoding='utf-8-sig') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['City','Name','Nosah','Address'])
        writer.writerows(output_rows)

    driver.quit()
    
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
cities = get_cities(driver)
nosahim = get_nosahim(cities,driver)

# print(cities)
```
